```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import the Data Result
data = pd.read_csv('result.csv')
```

```python
# Credit Score Segment Analysis
credit_score_df = data['credit_score']

defaultRate_bins = [519, 599, 649, 699, 749, 850]

defaultRate_labels = ['520-599', '600-649', '650-699', '700-749', '750+']

data['credit_score_range'] = pd.cut(credit_score_df, bins=defaultRate_bins, labels=defaultRate_labels)

default_rates = (data.groupby('credit_score_range')['defaulted'].mean() *100).sort_values(ascending=False)

print(default_rates)
```

```python
# Default Rate by Credit Score Range Visualization(Bar)
credit_scores_df = default_rates.reset_index()

plt.figure(figsize=(8,5))

plt.bar(credit_scores_df['credit_score_range'], credit_scores_df['defaulted'], color='royalblue')


plt.xlabel('Credit Score Range')
plt.ylabel('Default Rate (%)')
plt.title('Default Rate Skyrockets as Credit Score Drops')

plt.show()
```

```python
# DTI Ranges Segment Analysis
dti_df = data['dti_ratio']

dti_bins = [0,35,49,100]

dti_labels = ['looking good(0-35)', 'opportunity to improve(36-49)', 'take action(50+)']

data['dti_range'] = pd.cut(dti_df, bins=dti_bins, labels=dti_labels)

dti_range = (data.groupby('dti_range')['defaulted'].mean() * 100).sort_values(ascending=False)

print(dti_range)
```

```python
# DTI vs Default Status Visualization(Scatter Plot)
plt.figure(figsize=(8,5))

plt.scatter(data['dti_ratio'], data['defaulted'], alpha=0.2, color='royalblue')

plt.xlabel('Debt-to-Income Range')
plt.ylabel('Default Rate(%)')
plt.title('DTI vs Default Status')

plt.show()
```

```python
# Employment Status Segment Analysis
employment_status_range = (data.groupby('employment_status')['defaulted'].mean() * 100).sort_values(ascending=False)

print(employment_status_range)
```

```python
# Employment Year Segment Analysis
years_employed_df = data['years_employed']

yearsEmployed_bins = [0,2,50]

yearsEmployed_labels = ['< 2 Years', '2+ Years']

data['emp_length_group'] = pd.cut(years_employed_df, bins=yearsEmployed_bins, labels=yearsEmployed_labels)

emp_length = (data.groupby('emp_length_group')['defaulted'].mean() *100).sort_values(ascending=False)

print(emp_length)
```

```python
# Loan Purpose Segment Analysis
loan_purpose_range = (data.groupby('loan_purpose')['defaulted'].mean() * 100).sort_values(ascending=False)

print(loan_purpose_range)
```

```python
# Check Average loan amount on Default Status
average_loan_sizes = data.groupby('defaulted')['loan_amount'].mean()

print(average_loan_sizes)
```

```python
# Loan Purpose vs Default Rate Visualization(Bar)
purpose_rate = (data.groupby('loan_purpose')['defaulted'].mean() * 100).sort_values(ascending=False)

purpose_rate_df = purpose_rate.reset_index()

plt.figure(figsize=(10,8))

plt.barh(purpose_rate_df['loan_purpose'], purpose_rate_df['defaulted'], color='royalblue')

plt.xlabel('Default Rate(%)')
plt.ylabel('Loan Purpose')
plt.title('Loan Purpose vs Default Rate')

plt.show()
```

```python
# Correlation Analysis
numeric_columns = ['credit_score', 'dti_ratio', 'annual_income', 'loan_amount', 'interest_rate', 'defaulted']

correlation = data[numeric_columns].corr()

default_correlation = correlation['defaulted'].sort_values(ascending=False)

print(default_correlation)
```
